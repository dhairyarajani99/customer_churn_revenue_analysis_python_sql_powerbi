```python
mode_val = df['Returns'].mode()[0]
```

```python
df['Returns'] = df['Returns'].fillna(mode_val)
df['Returns'] = df['Returns'].astype(int)
```

```python
df.isnull().sum()
```

The `Purchase_Date` column is currently an object (string). We convert it to `datetime` format

```python
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'], dayfirst=True, format='mixed')
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop(columns=['Age'], inplace=True)
```

```python
df.columns
```

**Transaction_Value:** Created by multiplying `Product_Price` and `Quantity` to represent the actual revenue generated per row

```python
df['Transaction_Value'] = df['Product Price'] * df['Quantity']
```

```python
df.rename(columns={'Total Purchase Amount': 'Lifetime_Spend'}, inplace=True)
df[['Product Price', 'Quantity', 'Transaction_Value', 'Lifetime_Spend']].head()
```

**Month & Year Extraction:** We extract the month and year from `Purchase_Date`. This allows us to track seasonality, monthly sales growth, and yearly churn trends

```python
df['Month'] = df['Purchase Date'].dt.month
```

```python
df['Year'] = df['Purchase Date'].dt.year
```

**Age_Group:** Customers are segmented into demographic bins (Teen, Young Adult, Adult, Senior, Old) to analyze spending patterns by life stage

```python
bins = [0, 18, 30, 45, 60, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Senior', 'Old']

df['Age Group'] = pd.cut(df['Customer Age'], bins=bins, labels=labels)
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df = df.sort_values(by=['Customer_ID','Purchase_Date'])
df[df['Customer_ID'] == 46251]
```

**Key Finding:** Sorting revealed that `Lifetime_Spend` does not always increase chronologically. Because of this inconsistency, we will rely on our calculated `Transaction_Value` for revenue analysis
